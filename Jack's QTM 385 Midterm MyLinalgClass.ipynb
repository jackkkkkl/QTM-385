````python
# Your answers here

class Matrix:
    """
    Matrix Class
    
    Great to do Linear Algebra!
    """
    # Constructor
    def __init__(self, num, nrow):
        """
        Constructor:
        Receives:
            num: list of numbers
            nrow: number of rows
        
        Return: 
            Matrix object
        
        """
        self.len = len(num)
        self.nrow = nrow
        self.ncol = self.len // self.nrow
        
        self.mat = []
        l1 = 0
        for i in range(self.nrow):
            l2 = []
            for j in range(self.ncol):
                l2.append(num[l1])
                l1 += 1
            self.mat.append(l2)
            
        self.dim = [self.nrow, self.ncol]
        
        if self.len % self.nrow != 0:
            raise ValueError('Error. It is not a matrix.')
        else:
            pass
    
    # Print represetation
    def __str__(self):
        """
        Prints the matrix when using the command print()
        """
        pri=''
        for i in range(self.dim[0]):
            pri += str(self.mat[i])
            pri += "\n"
            
        return pri
        pass
    
    # Representation
    def __repr__(self):
        """
        __repr__:
            Prints the object description
            In this case, the matrix
        """
        return "Matrix dimension {}".format(self.dim)
        pass
    
    # Test to see if square matrix
    def is_square(self):
        """
        is_square:
            Receives:
                Matrix
                
            Return:
                True if the matrix is square
                False otherwise
        """
        if self.dim[0] == self.dim[1]:
            return True
        else:
            return False
        pass
        
    # Addition (see https://en.wikipedia.org/wiki/Matrix_addition)
    def __add__(self, other):
        """
        __add__:
            Implements the sum of matrices
            
            Receives:
                Two matrices
                
            Returns
                Subtraction of matrices as a Matrix or error if not conform
        """
        matr = []
        for i in range(self.dim[0]):
            add = []
            for j in range(self.dim[1]):
                add.append(self.mat[i][j])
            matr.append(add)
        if self.dim != other.dim:
            raise ValueError('Error. The matrices do not conform.')
        else:
            for i in range(self.dim[0]):
                for j in range(self.dim[1]):
                    matr[i][j] += other.mat[i][j]
            addin = []
            for i in range(self.dim[0]):
                for j in range(self.dim[1]):
                    addin.append(matr[i][j])
        return Matrix(addin, self.dim[0])
        pass
    
    # Subtraction
    def __sub__(self, other):
        """
        __sub__:
            Implements the subtraction of two matrices
            
            Receives:
                Two matrices
                
            Returns:
                Subtraction of matrices as a Matrix or error if not conform
        """
        matr = []
        for i in range(self.dim[0]):
            sub = []
            for j in range(self.dim[1]):
                sub.append(self.mat[i][j])
            matr.append(sub)
        if self.dim != other.dim:
            raise ValueError('Error. The matrices do not conform.')
        else:
            for i in range(self.dim[0]):
                for j in range(self.dim[1]):
                    matr[i][j] -= other.mat[i][j]
            subt = []
            for i in range(self.dim[0]):
                for j in range(self.dim[1]):
                    subt.append(matr[i][j])
        return Matrix(subt, self.dim[0])
        pass
        
    # Scalar multiplication (https://en.wikipedia.org/wiki/Scalar_multiplication)
    def __mul__(self, num):
        """
        __mul__:
            Implements scalar multiplication
            
            Receives:
                Matrix
                float or int
                
            Returns:
                Matrix
        """
        matr = []
        for i in range(self.dim[0]):
            mul = []
            for j in range(self.dim[1]):
                mul.append(self.mat[i][j])
            matr.append(mul)
        for i in range(self.dim[0]):
            for j in range(self.dim[1]):
                matr[i][j] *= num
        multi = []
        for i in range(self.dim[0]):
            for j in range(self.dim[1]):
                multi.append(matr[i][j])
        return Matrix(multi, self.dim[0])
        pass
    
    # Extract row
    def exrow(self, i):
        """
        exrow:
            Extracts row of a Matrix
            
            Receives:
                Matrix
                Index (int)
                
            Returns:
                List with row or error if index out of bounds
        """
        l = []
        if i >= self.dim[0]:
            raise ValueError("Error. Invalid row.")
        else:
            for j in range(self.dim[1]):
                l.append(self.mat[i][j])
            return l
        pass
    
    # Extract column
    def excol(self, i):
        """
        exrow:
            Extracts column of a Matrix
            
            Receives:
                Matrix
                Index (int)
                
            Returns:
                List with column or error if index out of bounds
        """
        l = []
        if i >= self.dim[1]:
            raise ValueError("Error. Invalid column.")
        else:
            for j in range(self.dim[0]):
                l.append(self.mat[j][i])
            return l
        pass
    
    # Matrix multiplication (see https://en.wikipedia.org/wiki/Matrix_multiplication)
    def __pow__(self, other):
        """
        __pow__:
            Implements the product of two matrices
            
            Receives:
                Two Matrices
                
            Returns:
                Matrix or error if not conform
        
        """
        if self.dim[1] != other.dim[0]:
            raise ValueError("Error. Matrices do not conform for multiplication.")
        else:
            l1 = []
            for i in range(self.dim[0]):
                l2 = []
                for j in range(other.dim[1]):
                    l3 = self.exrow(i)
                    l4 = other.excol(j)
                    a = 0
                    for k in range(self.dim[1]):
                        a += l3[k] * l4[k]
                    l2.append(a)
                l1.append(l2)
            l5 = []
            for i in range(self.dim[0]):
                for j in range(other.dim[1]):
                    l5.append(l1[i][j])
            return Matrix(l5, self.dim[0])
        pass
    
    # Extract submatrix for minor (see https://en.wikipedia.org/wiki/Minor_(linear_algebra))
    def subm(self, row, col):
        """
        subm:
            Extracts a submatrix for minor and cofactor computations
            
            Receives:
                Matrix
                row (int)
                col (int)
                
            Returns:
                Matrix
        """
        l = []
        for i in range(self.dim[0]):
            if i == row:
                continue
            for j in range(self.dim[1]):
                if j == col:
                    continue
                l.append(self.mat[i][j])
        return Matrix(l, self.dim[0] - 1)
        pass
    
    # Determinant (see https://en.wikipedia.org/wiki/Determinant)
    def det(self):
        """
        det:
            Compute the determinant of a matrix by cofactor expansion
            
            Receives:
                Matrix
                
            Returns:
                float, int or an error if matrix is not square.
        """
        def de(self):
            if self.dim[0] == 2:
                return self.mat[0][0] * self.mat[1][1] - self.mat[0][1] * self.mat[1][0]
            else:
                a = 0
                b = 0
                for i in range(self.dim[1]):
                    a += pow(-1, b+i) * self.mat[b][i] * de(self.subm(b, i))
                return a
        if self.dim[1] != self.dim[0]:
            raise ValueError("Error. Matrix is not square.")
        else:
            a = de(self)
            return a
        pass
    
    # Transpose
    def t(self):
        """
        t:
            Computes the transpose of a matrix
            
            Receives:
                Matrix
                
            Returns:
                Matrix
        """
        l = []
        for i in range(self.dim[1]):
            l += self.excol(i)
        return Matrix(l, self.dim[1])
        pass
    
    # Cofactor matrix (see https://en.wikipedia.org/wiki/Minor_(linear_algebra))
    def cofm(self):
        """
        cofm:
            Computes the Cofactor Matrix
            
            Receives:
                Matrix
                
            Returns:
                Matrix or error if matrix not square.
        """
        l = []
        for i in range(self.dim[0]):
            for j in range(self.dim[1]):
                l.append(pow(-1, i + j) * self.subm(i, j).det())
        return Matrix(l, self.dim[0])
        pass
    
    # Inverse matrix (see the inverse definition after buiding the cofactor matrix: 
    #                 https://en.wikipedia.org/wiki/Minor_(linear_algebra))
    def inv(self):
        """
        inv:
            Computes Inverse Matrix using cofactor matrix
            
            Receives:
                Matrix
                
            Returns:
                Matrix or error if matrix is singular or not square.
        """
        l = []
        a = self.cofm().t()
        b = self.det()
        for i in range(self.dim[0]):
            for j in range(self.dim[1]):
                l.append(a.mat[i][j] / b)
        return Matrix(l, self.dim[0])
        pass
````

````python
def cal(X, y):
    B = (X.t() ** X).inv() ** X.t() ** y
    return B
````
